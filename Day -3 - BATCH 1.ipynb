```python

def guessing_game():
  number_to_guess = random.randint(1,100)
  guess = None
  attempts = 0

  print("Welcome to the game!")
  print("I have selected a numebr from 1-100 . Can you guess it!")

  while guess != number_to_guess:
    guess = int(input("Enter your guess: "))
    attempts += 1

    if guess < number_to_guess:
      print("Too low! Try again")
    elif guess > number_to_guess:
      print("Too high ! Try again")
    else:
      print(f"CongratS you guessed in {attempts} attempts")

```

How do you generate a random string of a specified length consisting of uppercase letters and digits using the random module?

```python
full = "ABCDEFGHIJKLMNOPQRSTUVQXYZ0123456789"

def get_password(n):

  res = []
  for i in range(n):
    idx = random.randint(0,35)
    res.append(full[idx])

  return "".join(res)

get_password(5)
```
