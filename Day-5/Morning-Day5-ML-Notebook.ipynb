## Convert Categorical Variables

```python
display_all(df.info())
```

# remove missing values columns 

```python
df = df.dropna(thresh = len(df)*0.7, axis = 1)
df.info()
```

# fill missing values 

```python
df.LotFrontage.info()
```

```python
median_value = df['LotFrontage'].median()
df['LotFrontage'] = df['LotFrontage'].fillna(median_value)
```

```python
int_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[int_cols] = df[int_cols].fillna(df[int_cols].median())
df.info()
```

```python
df[['LotFrontage','SalePrice']].median()
```

```python
int_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[int_cols] = df[int_cols].fillna(df[int_cols].median())
df.info()
```

```python
obj_cols = df.select_dtypes(include=['object']).columns
# print(obj_cols)
df[obj_cols] = df[obj_cols].fillna(method='ffill')
df.info()
```

```python
obj_cols = df.select_dtypes(include = ['object']).columns
```

```python
# How to delete a column 
# df.drop(column_name, axis =1 )
# remove all the original object columns from the dataframe.
df = df.drop(obj_cols, axis = 1)
df.info()
```

```python
# now append the data to the original dataframe 
df = pd.concat([df, ohe], axis = 1)
display_all(df.isnull().sum())
```

## Splitting Data
- Split the dataset into training and validation sets.

```python
from sklearn.model_selection import train_test_split
```

```python
df.drop('Id',axis = 1, inplace=True)
df.head()
```

```python
y = df['SalePrice']
X = df.drop('SalePrice', axis = 1)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_valid.shape)
print(y_valid.shape)
```

## Model Selection and Training
- Choose a machine learning model and train it on the training data.

```python
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression()
model.fit(X_train,y_train)
lr_predict = model.predict(X_valid)
```

```python
y_valid - y_predict
```

```python
y_predict
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
rf_predict=model.predict(X_valid)
```

## Model Evaluation
- Evaluate the model on the validation set using appropriate metrics.
- Display the evaluation results.

```python
from sklearn.metrics import mean_squared_error

lr_error = mean_squared_error(y_valid, lr_predict)
rf_error = mean_squared_error(y_valid, rf_predict)

print(lr_error, rf_error)

import math 

print(f"Erorrs : {math.sqrt(lr_error)}\n {math.sqrt(rf_error)}")
```
